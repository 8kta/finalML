## Model
We make a model for first operation

```python
def nnain_model():
    model = Sequential()
    model.add(Dense(32, input_dim=2, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(16, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(loss='mean_squared_error',optimizer='adam',metrics=['binary_accuracy'])
    return model

model = nnain_model()
```

```python
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

#modelo.fit para la primera operación
target_data_op_a = np.array([[1],[1],[1],[0]], "float32")
model.fit(training_data, target_data_op_a, epochs=1000)
scores = model.evaluate(training_data,target_data_op_a)
```

then we want the predictions

```python
#a&b

pred_a = [[1.001], [0], [.001], [1]]
pred_b = [[0], [1], [0], [1]]
pred_c = [[0], [1], [1], [0]]

          
vectors_data = np.array([[pred_a[0][0],pred_b[0][0]],
                         [pred_a[1][0],pred_b[1][0]],
                         [pred_a[2][0],pred_b[2][0]],
                         [pred_a[3][0],pred_a[3][0]]], "float32")
predicted_value = model.predict(vectors_data)


print('The prediction of vector', vectors_data, 'is:\n')

for a in predicted_value:
    print(a)
```

we make the another operation's fitting and the composition with the last prediction

```python
#(a&b)$c
#model.fito para la primera operación
target_data_op_b = np.array([[0],[0],[0],[1]], "float32")
model.fit(training_data, target_data_op_b, epochs=1000)
scores = model.evaluate(training_data,target_data_op_b)

vectors_data = np.array([[predicted_value[0][0],pred_c[0][0]],
                         [predicted_value[1][0],pred_c[1][0]],
                         [predicted_value[2][0],pred_c[2][0]],
                         [predicted_value[3][0],pred_c[3][0]]], "float32")
predicted_value_b = model.predict(vectors_data)
```
