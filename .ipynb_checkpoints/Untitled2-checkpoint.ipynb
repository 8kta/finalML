La ecuación del cono que se usara en este problema es 
$$\frac{x^2+y^2}{c^2}=(z-z_0)^2$$
en donde $c=\frac{r}{h}$ en donde $h$ representa la altura del cono, $r$ el radio de la base del cono y $z_0$ es el vertice del cono. 
Luego entonces la ecuación del cono está dada por 
$$f(x,y)=\pm\frac{h}{r}\sqrt{x^2+y^2}+z_0.$$

```python
alt = input('Height of the cone: \n')
radius = input('Radius of the base: \n')
vertex = input('Vertex of cone:\n')

alt = float(alt)
radius = float(radius)
vertex = float(vertex)
```

```python
def f(x,y):
    
        return alt/radius*(x**2+y**2)**(1/2)+vertex # + stats.norm.rvs(loc=0,scale= .01,size=len
```

```python
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
#R = np.sqrt(X**2 + Y**2)
#Z = np.cos(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, f(X,Y), cmap='viridis',
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1, 13)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=2)

plt.show()
```

```python
# list of random float between a range 50.50 to 500.50
randomFloatList_x = []
# Set a length of the list to 10
for i in range(0, 100):
    # any random float between 50.50 to 500.50
    # don't use round() if you need number as it is
    x_data = round(random.uniform(-10, 10), 4)
    randomFloatList_x.append(x_data)
    
randomFloatList_y = []
# Set a length of the list to 10
for i in range(0, 100):
    # any random float between 50.50 to 500.50
    # don't use round() if you need number as it is
    y_data = round(random.uniform(-10, 10), 4)
    randomFloatList_y.append(y_data)
```

```python
X_data = pd.DataFrame(randomFloatList_x )
Y_data = pd.DataFrame(randomFloatList_y )
```

```python
#X_part = np.linspace(-10, 10, 200)
#Y_part = np.linspace(-10, 10, 200)

#new_list_x = X_part[:]
#new_list_y = Y_part[:]

#x_data = random.shuffle(new_list_x.reshape(-1,1))
#y_data = random.random(new_list_y.reshape(-1,1))

df = pd.DataFrame(np.column_stack(( X_data , Y_data , 
                                   f(X_data,Y_data))), 
                  columns=["X", "Y", 'Z'])

df
```

```python
def regression_model():
    model = Sequential()
    model.add(Dense(20, activation='relu', input_dim = 2 ))
    model.add(Dense(40, activation='relu'))
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = regression_model()
train_df = pd.DataFrame({'X' : randomFloatList_x, 'Y' : randomFloatList_y})
zeta = pd.DataFrame({'Z' : df['Z']})
train_df
```

```python
resultados = model.fit( train_df ,  zeta, validation_split=0.3, epochs=100, verbose=2)
```

```python
plt.plot(resultados.history['loss'])
plt.plot(resultados.history['val_loss'])
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Entrenamiento', 'Validación'], loc='upper left')
plt.show()
```
